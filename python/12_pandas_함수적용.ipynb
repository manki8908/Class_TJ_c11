```python
import seaborn as sns
df = sns.load_dataset('titanic')
print(df.head())
print(df.shape)
```

```python
# age를 평균으로 채워 넣기
df['age'].fillna(int(df['age'].mean(axis=0)), inplace=True)
df.isnull().sum()
```

```python
# embarked을 최빈값으로 대체
most_freq = df['embarked'].value_counts(dropna=True).idxmax()  # return idex
print(most_freq)
df['embarked'].fillna(most_freq, inplace=True)
df.isnull().sum()
```

```python
# embark_town 결측값이 있는 행 삭제
df.dropna(subset=['embark_town'], axis=0, inplace=True)
df.isnull().sum()
df.shape
```

```python
# deck 삭제
# embark_town 결측값이 있는 행 삭제
df.drop(columns=['deck'], axis=1, inplace=True)
df.isnull().sum()
df.head()
```

```python
# titanic['adult/child'] 컬럼을 만들고 20살 이상이면 adult 이하면 child
# apply, lambda 활용
import numpy as np

def adult_child(age):
    if age >= 20:
        man='adult'
    else:
        man='child'
    return man

df['adult/child'] = np.nan
df['adult/child'] = df.apply(lambda x: adult_child(x['age']), axis=1)
#df['adult/child'] = df.age.apply(lambda x: 'adult' if x > 20 else 'child')
df.tail()
```

```python
# bins=[1,20,30,50,70,100]
# label = ['미성년자', '청년', '중년', '장년', '노년']
# age를 기준으로 'age_cut'이라는 컬럼 생성후 두 컬럼을 출력하기, (cut 함수)
# 22 청년
# 38 주년

bin_divers = [1,20,30,50,70,100]
# 3개의 bin에 이름지정
bin_name = ['미성년자', '청년', '중년', '장년', '노년']

# cut 함수로 각데이터를 3개의 bin에 할당
df['age_cut'] = pd.cut(x=df['age'], 
                      bins=bin_divers,
                      labels=bin_name,
                      include_lowest=True)     # 첫경계값의 포함 여부
print(df[['age', 'age_cut']].head(10))
```
