```python
student1 = pd.Series(dict(국어=100, 영어=80, 수학=90))
student2 = pd.Series(dict(국어=90, 영어=80, 수학=80))
```

```python
# 두학생의 과목별 점수 사칙연산
add = student1 + student2
add1
```

```python
sub = student1 - student2
```

```python
mul = student1 * student2
```

```python
div = student1 / student2
```

```python
# 사칙연산 결과를 데이터 프레임으로 합치기
# 여러개 시리즈를 프레임으로
result = pd.DataFrame([add,sub,mul,div], index=['덧셈','뺄셈','곱셈','나눗셈'])
result
```

```python
# 결측이 있을시, nan값이 있는 곳은 연산이 안됨
student1 = pd.Series(dict(국어=np.nan, 영어=80, 수학=90))
student2 = pd.Series(dict(국어=90, 영어=80, 수학=80))

add2 = student1 + student2
sub2 = student1 - student2
mul2 = student1 * student2
div2 = student1 / student2

result = pd.DataFrame([add2,sub2,mul2,div2], index=['덧셈','뺄셈','곱셈','나눗셈'])
result
```

```python
# 두학생의 과목별 점수 사칙연산(연산메소드 사용)
add3 = student1.add(student2, fill_value=0)
sub3 = student1.sub(student2, fill_value=0)
mul3 = student1.mul(student2, fill_value=0)
div3 = student1.div(student2, fill_value=0)
print(add3,sub3,mul3,div3)
```
