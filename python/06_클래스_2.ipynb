```python
class TunaCanSet():

    def __init__(self, normal, veg, pepper):
        self.normal = normal
        self.veg = veg
        self.pepper = pepper
        print('TunaCanSet initialize')

    def content_show(self, name):
        print('---------', name, '---------')
        print('Normal Tuna: ', self.normal)
        print('Vegitable Tuna: ', self.veg)
        print('Pepper Tuna: ', self.pepper)

tunaset01 = TunaCanSet(12,5,6)
tunaset01.content_show('Tuna GiftSet 01')
```

#상속

```python
class 세면도구():
 
    def __init__(self, 치약, 샴푸, 비누):
        self.치약 = 치약
        self.샴푸 = 샴푸
        self.비누 = 비누

    def 내용물출력(self, 세트이름):
        print('**', 세트이름, '**')
        print('일반참치', self.치약)
        print('야채참치', self.샴푸)
        print('고추참치', self.비누)

class 특별세트(세면도구):    # 상속

    def __init__(self, 치약, 샴푸, 비누, 면도기):
        super().__init__(치약, 샴푸, 비누) # 오버라이드
        self.면도기 = 면도기

    def 내용물출력(self, 세트이름):
        super().내용물출력(세트이름)       # 오버라이드
        print('면도기', self.면도기)

특01호 = 특별세트(1,2,3,1)
특01호.내용물출력('명절 특별세트 01호')
```

# 클래스 복습
자동차 클래스와 객체 \
* 속성: 브랜드, 모델, color 
* 메소드: drive, turn_off/on

```python
class 자동차():

    def __init__(self, 브랜드, 모델, 색):
        self.브랜드 = 브랜드
        self.모델 = 모델
        self.색 = 색
        print("----- 출고 ------")
        print("브랜드: ", self.브랜드)
        print("모델: ", self.모델)
        print("색 ", self.색)
             
    def 기능(self,시동):
        self.시동 = 시동
        print("시동: ", self.시동)
        
    def 운전모드(self,모드):
        self.모드 = 모드
        print("모드: ", self.모드)
        
자동차1 = 자동차('기아', '쏘렌토', '화이트')
자동차1.기능('켬')
자동차1.운전모드('전기')
```

```python
# car class를 만들고
# 속성: 이름, cc, year
# 기능: check - cc가 1000 미만이면 소형, 1000~2000 중형, 2000 이상 대형
# car 객체를 만들때 car_count 개수가 하나씩 증가하도록
# car1: 소나타, 2500, 2019 --> 대형, car_count = 1대
# car2: 아반테, 1500, 2018 --> 중형, car_count = 2대

class car():

    car_count = 0

    #def test():
    #    print(car.car_count)

    def __init__(self, 이름, cc, year):
        self.이름 = 이름
        self.cc = cc
        self.year = year
        car.car_count += 1
        print("----- 출고 ------")
        print("이름: ", self.이름)
        print("CC: ", self.cc)
        print("Year ", self.year)
        #print("출고량", car.car_count)

    def check_cc(self):
        if self.cc < 1000: print('소형')
        if 1000 <= self.cc < 2000: print('중형')
        if self.cc >= 2000: print('대형')

car1 = car('소나타', 2500, 2019)
car1.check_cc()
car1.car_count
car2 = car('아반테', 1500, 2018)
car2.check_cc()
car1.car_count

    
```

```python
# 1. Character 클래스 만들기
# 캐릭터의 속성값에 health 속성(생성시 200)    
# 캐릭터 클래스 move() 메소드, 사용시 health -10
# 캐릭터 클래스 rest() 메소드, 사용시 health +10
# 현재 health를 알 수 있는 checkHealth() method 추가
# 

class Character():

    def __init__(self, health=200):
        self.health = health
        print("캐릭터가 생성되었습니다")

    def move(self):
        self.health -= 10
        print("캐릭터가 이동하였습니다, HP -10")
    
    def rest(self):
        self.health += 10
        print("캐릭터가 휴식하였습니다, HP +10")

    def checkHealth(self):
        print("현재 체력은 ", self.health)

character1 = Character()
character1.checkHealth()
character1.move()
character1.rest()
character1.checkHealth()

# 2. Knight, Healer 클래스 만들기
# Character 클래스 상속받기
#
# Knight move()를 하면 health가 추가로 -5 더 소모하도록
# Knight Attack()를 추가하고 '공격합니다'라는 메소드 작성

class Knight(Character):

    def __init__(self, health=200):
        super().__init__(health)
        self.health = health
        print("Knight 입니다, 체력= ", self.health)

    def move(self):
        super().move()
        self.health -= 5
        print("knight는 -5가 더 깎입니다")

    def attack(self):
        print("knight가 공격합니다")

character2 = Knight()
character2.move()
character2.checkHealth()
character2.attack()

# Heaer Mana() 속성 추가 생성시 100의 값을 주어짐
#       heal(Character) 메소드를 추가하고 Character들을 매개변수로 받아
# heal(Character) 메소드를 실행하면 Mana가 -10이 되고, 전달받은 Character 객체의
# rest() 메소드가 실행되어 +10이 되도록 함
# Healer는 현재 마나 속성을 확인할 수 있는 checkMana() 메소드를 추가

class Healer(Character):

    def __init__(self, health=200, mana=100):
        super().__init__(health)
        self.health = health
        self.mana = mana
        print("Healer 입니다, 체력= ", self.health)
        print("             , 마나= ", self.mana)

    def heal(self, man):
        man.rest()
        self.mana -= 10
        print('마나를 10 사용하였습니다')

    def checkHealth(self):
        super().checkHealth()

    def checkMana(self):
        print("현재 마나는 ", self.mana)

character3 = Healer()
character3.heal(character2)
character3.checkHealth()
character3.checkMana()


```
