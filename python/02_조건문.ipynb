```python
# if문, 가위,바위,보 게임

scissor, rock, paper = '가위', '바위', '보'
srp = ['가위', '바위', '보']
win, lose, draw = '승리', '패배', '비김'

while 1:

    # .. Mine
    mine = input("고르세요: (게임 종료를 원하시면 q를 입력하세요)")
    if mine == 'q': break

    # .. yours
    rann = int(input('난수를 고르세요'))
    selc = rann%3
    yours = srp[selc]
    #yours = '바위'

    print(f'당신: {mine}, 상대: {yours}')

    if mine == yours:
        result = draw
    else:
        if mine == scissor:
            if yours == rock: result = lose
            if yours == paper: result = win
        elif mine == rock:
            if yours == scissor: result = lose
            if yours == paper: result = win
        elif mine == paper:
            if yours == scissor: result = lose
            if yours == rock: result = win

    print(result)
```

```python
# 업다운 게임

st, ed = 1, 50
chance = 3
stop = 21

while chance != 0:

    go = int(input("숫자를 고르세요"))

    if go == stop:
        print("정답!")
        break
    else:
        if go > stop:
            print("다운")
        if go < stop:
            print("업")
        chance += -1

    print(f'남은 기회: {chance}')
    if chance == 0: print('실패')


```
