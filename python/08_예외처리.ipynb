```python
# try, except

a = [1,2,3]
#a[3]

try:
    a[3]
except Exception as e:
    print(e)
```

```python
a = [1,2,3]

try:
    x = int(input())
    y = 100 / x
    print(y,a[x])
except Exception as e:
    print(e)
```
