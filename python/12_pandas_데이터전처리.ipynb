```python
# 결측치 확인 for 반복문으로 각 열의 Nan 개수 계산
missing_df = df.isnull()
for col in missing_df.columns:
    missing_count = missing_df[col].value_counts()
    try:
        print(col, ':', missing_count[True])   # nan 값이 있으면 개수를 출력
    except:
        print(col, ":", missing_count[0])      # nan 값이 없으면 0 출력
```

```python
# age 열에 나이데이터 없는 모든 행을 삭제()
#df_age = df_tresh.dropna(subset=['age'], how='any', axis=0)
df_age = df.dropna(subset=['age'], how='any', axis=0)
df_age.isnull().sum()
```

```python
# age 열의 nan 값을 다른 나이 데이터의 평균으로 대체
mean_age = df['age'].mean(axis=0)   # age 열의 평균을 계산(nan값은 제외)
mean_age
df['age'].fillna(int(mean_age), inplace=True)
df_age.isnull().sum()
```

```python
# embark_town(승선지)는 최빈값(가장 많이 출연한 값)으로 대체
most_freq = df['embark_town'].value_counts(dropna=True).idxmax()  # return idex
#most_freq = df['embark_town'][?:?].value_counts(dropna=True).idxmax()  # 구역 최빈값, return idex
print(most_freq)
df['embark_town'].fillna(most_freq, inplace=True)
df.isnull().sum()
```

```python
# embarked 열의 Nan값을 바로 앞에 있는 값으로 변경하기
df['embarked'].fillna(method='ffill', inplace=True)   # foward fill
df.isnull().sum()
```

```python
# .. 방법2, 원핫인코딩
# 1. 문자열 -> 라벨링: label_encoder
# 2. 라벨링 -> 원핫인코딩: onehot_labeled
from sklearn import preprocessing

# 전처리를 위한 enconder 객체 만들기
label_encoder = preprocessing.LabelEncoder()        # 범위에 대한 라벨링
onehot_encoder = preprocessing.OneHotEncoder()      # 라벨링 데이터 더미처리
```

```python
# label encoder
import seaborn as sns
import pandas as pd
import numpy as np

df = sns.load_dataset('mpg')
df.head(3)
```

```python
# 물음표 처리
df['horsepower'].replace('?', np.nan, inplace=True)
#print(df['horsepower'].unique())

# 결측제거
df.dropna(subset=['horsepower'], axis=0, inplace=True)
#print(df.isnull().sum())

df['horsepower'] = df['horsepower'].astype('float')

# 3개 범위 가젹오기
count, bin_divers = np.histogram(df['horsepower'], bins=3)
#print(bin_divers)

# 3개의 bin에 이름지정
bin_name = ['저출력', '보통출력', '고출력']

# cut 함수로 각데이터를 3개의 bin에 할당
df['hp_bin'] = pd.cut(x=df['horsepower'], 
                      bins=bin_divers,
                      labels=bin_name,
                      include_lowest=True)     # 첫경계값의 포함 여부

# label encoder로 문자열  범주를 숫자형 범주로 변환
onehot_labeled = label_encoder.fit_transform(df['hp_bin'].head(15))
onehot_labeled        # 보:1, 고:0, 저:2,       1차원 출력

# 2차원 행렬형태 변환
onehot_reshaped = onehot_labeled.reshape(len(onehot_labeled),1)
print(onehot_reshaped)
print(type(onehot_reshaped))
```

```python
onehot_fitted = onehot_encoder.fit_transform(onehot_reshaped)
print(onehot_fitted)
print(type(onehot_fitted))            
# 0번째,  1에  1   0 1 0
# 5번째,  0에  1   1 0 0
# 14번째, 0에  1   0 0 1
```

```python
dates = ['2019-01-01', '2020-03-01', '2021-06-01']
ts_dates = pd.to_datetime(dates)
print(ts_dates)
```

```python
#Timestamp를 period로 변환
pr_day = ts_dates.to_period(freq='D')
print(pr_day)
pr_month = ts_dates.to_period(freq='M')
print(pr_month)
pr_year = ts_dates.to_period(freq='A')
print(pr_year)
```

```python
# timestamp의 배열 만들기 - 월간격, 월의 시작일로
ts_ms = pd.date_range(start='2019-01-01',  # 시작일
                      end=None,            # 끝날짜 
                      periods=6,            # 생성할 timesamp의 개수
                      freq='MS',           # 시간 간격, MS: 월의 시작일
                      tz = 'Asia/Seoul')   # 시간대 지역
```
