```python
# 숫자를 입력받아서 짝수인지 홀수인지 구분하는 함수
# 짝수 even, 홀수 odd
# input: 10, output: 'even'

def even_odd(x):
    res = x%2
    out='odd'
    if res == 0: out='even'   
    return out

print(even_odd(10))
print(even_odd(7))
```

```python
# seven(num) 처럼 num을 입력받은 수를 초과해서 7의 배수중에서 제일 작은수를 찾는 함수
# seven(22) 28, 35, 42
#

def seven(num):     
    if num%7 != 0: 
        a = num//7
        c = (a+1)*7 
    else:
        c = num+7
    return c

print(seven(77))
```

```python
# 기초통계
from statistics import mean, variance
from math import sqrt

dataset = [2,4,5,6,1,8]


# 산술평균
def avg1(dataset):
    avg = mean(dataset)   # 언패킹 자동
    return avg
print('산술평균: ', avg1(dataset))


# 분산/표준편차
def var_sd(dataset):
    avg = avg1(dataset)
    dif = [ (d - avg)**2 for d in dataset ]
    var = sum(dif) / (len(dataset)-1)
    std = sqrt(var)
    return var, std

print(var_sd(dataset))

```

```python
# 동전던지기
import random
def coin(n):
    result = []
    for i in range(n):
        r = random.randint(0,1)
        if ( r == 1 ):
            result.append(1)
        else:
            result.append(0)     
    return result

# 몬테카를로
def montaCoin(n):
    cnt = 0
    for i in range(n):
        cnt += coin(1)[0]
    res = cnt / n
    return res

print(montaCoin(10))
print(montaCoin(30))
print(montaCoin(100))
print(montaCoin(1000))
print(montaCoin(10000))
print(montaCoin(100000))
```

```python
# print_5xn('string') 입력받아 한줄에 다섯글자씩만 출력하는 함수를 작성
# '아이엠어보이유알어걸  --> 아이엠어보
#                            이유알어걸

def print_5xn(data):
    b = []
    l = len(data) // 5
    for i in range(l):
        b.append(data[i*5:(i+1)*5])
    return b

print_5xn('아이엠어보이유알어걸')
```
