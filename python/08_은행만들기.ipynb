# 은행만들기

```python
# - 속성: 계좌번호: '123456' / owner: 이름 (문자열) / 잔액(balance), 비공개속성
# - method: 저금(deposit) 1000만원 이상 입금할 수 없음
# -         출금(withdraw), 잔액이 0이면 출금 안되도록
# -         조회

class Account():

    def __init__(self, ano, owner, balance):
        self.ano = ano
        self.owner = owner
        self.balance = balance

    def deposit(self, amount):
        if amount + self.balance >= 10000000:
            return '천만원 이상은 입금할 수 없습니다'
        self.balance += amount

    def withdraw(self, amount):
        if self.balance - amount < 0:
            return '잔액이 부족합니다'
        self.balance -= amount

    def 조회(self):
        print('잔액: ', self.balance)

    def __str__(self):         # __str__의 기능을 오버라이드
        return f'계좌번호:{self.ano}, 소유주:{self.owner}, 잔액{self.balance: 9,d}'
```

```python
test = Account('123456', '홍길동', 3000000)
test.deposit(3000000)
print(test)
```

```python
test.deposit(700000)
print(test)
```

```python
test.withdraw(700000)
print(test)
```

```python
test2 = Account('098904', '마리아', 3000000)
print(test2)
```

```python
acc_list = [test, test2]
def check_duplicate(ano):
    for acc in acc_list:
            if ano == acc.ano:
                return True
    return False
```

```python
# 사용자로부터 필요한 정보를 입력받아서 계좌를 생성
def create_account():
    while 1:
        s = input('계좌번호 성명 금액을 입력 > ').split()
        ano, owner = s[0], s[1]
        amount = int(s[2])
        if not check_duplicate(ano):
            break
    acc = Account(ano, owner, amount)
    acc_list.append(acc)
```

```python
# 사용자로부터 필요한 정보를 입력받아서 계좌에 돈을 입금
def deposit_account():
    s = input('계좌번호 금액 > ').split()
    ano, amount = s[0], int(s[1])
    for acc in acc_list:
        if acc.ano == ano:
            acc.deposit(amount)
            return
```

```python
# 사용자로부터 필요한 정보를 입력받아서 계좌에 돈을 출금
def withdraw_account():
    s = input('계좌번호 금액 > ').split()
    ano, amount = s[0], int(s[1])
    for acc in acc_list:
        if acc.ano == ano:
            acc.withdraw(amount)
            return
```

```python
# 화면
while 1:
    menu = int(input('1:계좌생성, 2:계좌목록, 3:입금, 4:출금, 5:종료 >'))
    if menu == 5: break
    if menu == 1: 
        create_account()
    elif menu == 2:
        for account in acc_list:
            print(account)
    elif menu == 3:
        deposit_account()
    elif menu == 4:
        withdraw_account()
    print()
    
```
