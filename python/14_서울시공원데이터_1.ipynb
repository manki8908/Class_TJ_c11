```python
import numpy as np
import pandas as pd
import folium
import re           # 정규표현식
```

```python
df = pd.read_csv('/content/서울시 주요 공원현황.csv', encoding='EUC-KR')
df.info()
```

```python
# 면적정보 추출
# apply? agg?
area = df.면적.astype(str).apply(lambda x: x.strip('총 ').split()[0])
area.unique()
# m2, m3 등 이상한 단위 포함
```

```python
area[ (area == 'nan') | (area == '휴양')] = '0㎡'
```

```python
area.unique()
```

```python
# 단위가 이상한것들 정규표현식으로 제거
# 실수로 변환후 값확인 후 정수로 최종 변환
area_sub = area.apply( lambda x: float(re.sub('[㎡㎥m2m²]', '',x)) )
area_sub = area_sub.apply(lambda x: int(np.round(x)))
#area_sub = area_sub.apply(int)
#df['면적'] = area_sub
```

```python
area_sub.unique()
```

# 공원의 면적에 따라 분류

```python
# 기준 설정
area_criteria = [-1, 100000, 1000000, 12000000]
labels = ['소형', '중형', '대형']
size_info = [3, 7, 15]

scale = pd.cut(area_sub, area_criteria, labels=labels, include_lowest=True)
size = pd.cut(area_sub, area_criteria, labels=size_info)

                    
#print(df[['horsepower', 'hp_bin']].head(10))
```

# 새로운 프레임 만들기

```python
df = df[['공원명', '지역', 'X좌표(WGS84)','Y좌표(WGS84)']]
df.columns=['공원명', '지역', '경도', '위도']
df.head()
```

```python
df['면적'] = area_sub
df['스케일'] = scale
df['크기'] = size
```

```python
df[df.경도.isna()]
```

```python
df = df.dropna()
```

```python
df.to_csv('서울공원요약.csv', index=False, encoding='cp949')
df
```
