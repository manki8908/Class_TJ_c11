```python
# 1. 딕셔너리명 = dict(키1=값1, 키2=값2, 키3=값3....)
# 주의 키에는 ''를 잘 쓰지 않음

menu1 = dict(김밥=3000, 떡볶이=2000, 어묵=3500, 튀김=2800)
print(menu1)
```

```python
# dict(zip(key,value))

key_list = ['김밥', '떡볶이', '어묵', '튀김']
value_list = [3000, 2000, 3500, 2800]
menu2 = dict(zip(key_list, value_list))
print(menu2)
```

```python
# dict( [ (키1,값1), (), () ] )
menu3 = dict( [ ('김밥',3000), ('떡볶이',2000), ('어묵',3500), ('튀김',2800) ] )
print(menu3)
```

```python
이름 = input('이름 : ')
나이 = input('나이 : ')
개인정보 = dict(이름=이름, 나이=나이)
for k, v in 개인정보.items():
    print('key', k, '  value', v)
```

```python
# 아이템 출력 2/ enumerate
for i, (k, v) in enumerate(개인정보.items()):
    print(i, '  key', k, '  value', v)
```

```python
import requests
btc = requests.get("https://api.bithumb.com/public/ticker/").json()['data']
btc
```

```python
std = float(btc['opening_price'])
max = float(btc['max_price'])
min = float(btc['min_price'])
gap = max - min
if std + gap > max: 
    print("상승장")
else:
    print("하락장")
```
