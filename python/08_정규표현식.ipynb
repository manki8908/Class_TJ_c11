```python
import re
```

```python
# '.'은 한개의 임의 문자임

r = re.compile("a.c")
r.search('kkk')
```

```python
r.search('abc')       # match 된 단어를 3개 찾음, '.' 임의의 문자라서
```

```python
# ?는 ? 앞에 문자가 존재할 수 도 있고, 않을 수도 있는 경우
r = re.compile("ab?c")  # --> abc or ac
r.search('abbc')   # 못찾음
```

```python
r.search('abc')
```

```python
r.search('ac')
```

```python
# *는 바로앞에 문자가 0개 이상일 경우
r = re.compile("ab*c")
r.search('ac')
```

```python
r.search('abbbbc')
```

```python
# +는 앞의 문자가 최소 1개 이상
r = re.compile('ab+c')
r.search('abc')
```

```python
# ^는 시작되는 문자열
r = re.compile('^ab')
r.search('zab')
```

```python
r.search('abz')
```

```python
# {숫자}, 해당문자를 숫자만큼 반복
r = re.compile('ab{2}c')
r.search('abbc')
```

```python
r = re.compile('ab{2,8}c')  #{숫자1,숫자2} 행문자를 숫자1이상, 숫자2 이하 만큼 반복: 2~8
r.search('abbbbbbbbc')
```

```python
# [] 안에 문자를 넣으면 그 문자들 중에서 한개의 문자와 매치하라는 의미
r = re.compile('[abc]')  # [a,b,c,....z], [a-z], [a-c]
r.search('a')
```

```python
r.search('ba')
```

```python
# 알파벳 소문자
r = re.compile("[a-z]")
r.search('aBC')
```

```python
# 대문자
r = re.compile("[A-Z]")
r.search('aBC')
```

```python
# 대문자,소문자
r = re.compile("[A-Za-z]")
r.search('aBC')
```

```python
# [^문자] ^뒤에 붙은 문자들을 제외한 모든 문자를 매치하라는 역
r = re.compile("[^abc]")
r.search('d')
```

# 정규포현식 교재 201P 문제 

```python
texts = ['우리나라, 대한민국, 우리나라%$ 만세',
         '위산과&다 500GRAM 소화력 최고!',
         '나는 대한민국 사람',
         '보험료 15000원에 평생 보장 마감 임박',
         '나는 홍길동']
```

```python
# 소문자로 변경
#b = [ i*3 for i in a ]
b = [ t.lower() for t in texts ]
b
```

```python
# 숫자제거
b = [ re.sub("[0-9]", '',text) for text in texts ]
b
```

```python
# 문장부호 제거[. , : ;]
b = [ re.sub("[,.?!:;]", '',text) for text in texts ]
b
```

```python
# 특수문자 제거[$%*()]
b = [ re.sub("[@#$%^&*()]", '',text) for text in texts ]
b
```

```python
# 영문자제거
b = [ ''.join(re.findall("[^a-z]", text)) for text in texts ]
b
```

```python
# 공백제거
b = [ re.sub(" ", '',text) for text in texts ]
b
```
