##산술연산자

```python
num1=100
num2=2.

add = num1 + num2
print('add= ', add)

sub = num1 - num2
print('sub= ', sub)

mul = num1 * num2
print('mul= ', mul)

div1 = num1 / num2
print('div1= ', div1)

div2 = num1 % num2
print('div2= ', div2)

div3 = num1 // num2    # 정수, 몫만 반환
print('div3= ', div3)

sqr = num1**num2
print('sqr= ', sqr)
```

#IF문

```python
money = 3000
if money >= 100000:
   print('저금을 한다.')
else:
   print('대출을 받는다')
```

```python
money = 2000
bitcoin = 1
if (money >= 3000) or bitcoin:
   print('저금을 한다.')
else:
   print('대출을 받는다')
```

# &, | bitwise 연산: and, or와 유사

```python
money = 2000
bitcoin = 1
if money <= 3000 & bitcoin:
   print('저금을 한다.')
else:
   print('대출을 받는다')
```

```python
money = 2000
bitcoin = 1
if money >= 3000 | bitcoin:
   print('저금을 한다.')
else:
   print('대출을 받는다')
```

```python
result = 7 & 2
print(result)
```
