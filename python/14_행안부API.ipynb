```python
# 온라인으로 보내거나 받아올 때 사용, 파이썬으로 웹사이트 연동할 때
import requests

# url에서 한글 코드
# %EC%84%9C%EC%9A%B8%ED%8A%B9%EB%B3%84%EC%8B%9C%EC%B2%AD = '서울특별시청'
from urllib.parse import quote
```

```python
# 네이버에서 서울특별시청 검색 주소 복붙 내용
# 사이트에서 보이는거랑 달리 문자열로 보임
# https://search.naver.com/search.naver?where=nexearch&sm=top_sug.pre&fbm=0&acr=4&acq=%EC%84%9C%EC%9A%B8%ED%8A%B9%EB%B3%84%EC%8B%9C%EC%B2%AD&qdt=0&ie=utf8&query=%EC%84%9C%EC%9A%B8%ED%8A%B9%EB%B3%84%EC%8B%9C%EC%B2%AD
```

```python
# 요청 정보 url 작성
#ex json) https://business.juso.go.kr/addrlink/addrLinkApi.do?currentPage=1&countPerPage=10&keyword=강남대로12길&confmKey=TESTJUSOGOKR&resultType=json
bldg = '서울특별시청'
temp = quote(bldg)
base_url = 'https://business.juso.go.kr/addrlink/addrLinkApi.do'
params1 = f'currentPage=1&countPerPage=10&keyword={temp}'
params2 = f'&confmKey={api_key}&resultType=json'
url = f'{base_url}?{params1}?{params2}'
#print(url)
```

```python
# 정보 요청
# text
#result = requests.get(url)
#print(result.status_code)
#result.txt

# json
result = requests.get(url).json()
result
```

```python
road_add = result['results']['juso'][0]['roadAddr']
road_add
```

# 5지점 정보 가져오기 예

```python
places = '종로구청 노원구청 송파구청 마포구청 양천구청'.split(' ')
places
```

```python
base_url = 'https://business.juso.go.kr/addrlink/addrLinkApi.do'
params2 = f'&confmKey={api_key}&resultType=json'
road_add_list = []
for i in places:
    temp = quote(i)
    params1 = f'currentPage=1&countPerPage=10&keyword={temp}'
    url = f'{base_url}?{params1}?{params2}'
    result = requests.get(url).json()
    road_add_list.append(result['results']['juso'][0]['roadAddr'])

road_add_list
```

```python
import pandas as pd
df = pd.DataFrame(dict(공공기관=places, 도로명주소=road_add_list))
df.to_csv('공공기관_행안부.csv', encoding="cp949")
```
