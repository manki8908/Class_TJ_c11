리스트(튜플) 통계값 구하기
* 항목수: len()
* 합계: sum()
* 최소값: min()
* 최대값: max()
* 항목의개수: 리스트.count(항목)

```python
(tp) = (1,2,3,4,5,3,4,5)
print(len(tp))
print(sum(tp))
print(min(tp))
print(max(tp))
print(tp.count(3))
print( sum(tp)/len(tp) )
```

#set 집합
- 교집합
& or intersection
- 합집합
| or union

* &,|: 집합에선 비트연산 아님

```python
s1 = set([1,2,3])
s1[0]  # set은 indexing이 안됨
```

```python
s2 = set('hello')
s2  # 중복을 허용안함
```

```python
# 교집합
s1 = set([1,2,3,4,5,6])
s2 = set([4,5,6,7,8,9])
s1 & s2
```

```python
s1.intersection(s2)
```

```python
# 합집합
print(s1|s2)
print(s1.union(s2))
```

```python
# 차집합
print(s1-s2)
print(s1.difference(s2))
print(s2.difference(s1))
```

```python
# 추가
s3 = set([1,2,3])
s3.add(4)
s3
```

```python
# 추가
s3.update([5])
s3
```

```python
s3.remove(2)   # 없으면 에러남
s3.discard(2)  # 없어도 에러가 안남
```

```python
s4 = {"1", 3,5,(1,3)}  # 집합은 순서가 없음
s4
```

```python
s5 = {"1", 3,5,[1,3]} # 리스트 안됨(mutable 변화)
```

```python
s6 = {"1", 3,5,{1,3}} # set 안됨(mutable 변화)
```

```python
s7 = {"1", 3,5,frozenset([1,3])} # 함수를 넣어주면 변화시킬수 있긴 있음
```

#리스트 내포(list comprehension)

```python
a = [1,2,3,4]

# [3,6,9,12] 출력
b = []
for i in a:
    b.append(i*3)
print(b)
```

```python
# 내포
b = [ i*3 for i in a ]
b
```

```python
# 3을 곱하고 짝수만 출력 
c = []
for i in a:
    if i*3%2 == 0:
        c.append(i*3)
c
```

```python
a = [1,2,3,4]
c = [ i*3 for i in a if i*3%2 == 0 ]
c
```

```python
# 구구단 코드 내포 [2,4,6,8,]

c = [ i*y for i in range(2,10) for y in range(1,10) ]
c
```
