안쓰면 인스턴스끼리 공용

```python
class 세면도구:
    치약 = 0
    샴푸 = 0
    비누 = 0
```

```python
세01호 = 세면도구()  # 인스턴스
세02호 = 세면도구()  # 인스턴스
세03호 = 세면도구()  # 인스턴스

세01호.치약 = 12
세01호.샴푸 = 3
세01호.비누 = 3
```

```python
print('세면도구 01호 선물세트 내용물')
print('치약: ', 세01호.치약, '샴푸: ', 세01호.샴푸, '비누: ', 세01호.비누, sep=" ")
```

```python
class 참치선물세트():
    일반 = 0
    야채 = 0
    고추 = 0

    def 총합(self, 이름):
        합 = self.일반 + self.야채 + self.고추
        return 이름 + str(합)

참01호 = 참치선물세트()
참01호.일반 = 10
참01호.야채 = 5
참01호.고추 = 2

참01호.총합('참치선물세트 01호의 내용물: ')
```

```python
class 참치선물세트():
    일반 = 0
    야채 = 0
    고추 = 0

    def 출력(self, 이름):
        print('**', 이름, '**')
        print('일반참치', self.일반)
        print('야채참치', self.야채)
        print('고추참치', self.고추)

참03호 = 참치선물세트()
참03호.일반 = 7
참03호.야채 = 5
참03호.고추 = 10

참03호.출력('참치선물세트 03호 내용물')
        
```

```python
class 참치선물세트():
    일반 = 0
    야채 = 0
    고추 = 0

    def 출력(self, 이름):
        print('**', 이름, '**')
        print('일반참치', self.일반)
        print('야채참치', self.야채)
        print('고추참치', self.고추)

    def 출력2(self):
        self.출력('참치선물세트')

참04호 = 참치선물세트()
참04호.일반 = 7
참04호.야채 = 5
참04호.고추 = 10

참04호.출력2()
```
