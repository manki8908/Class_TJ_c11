```python
# r=raw 이스케이프문자, 
file = glob(r'C:\workspace\VScode\crawling\dain\주유소/지역*.xls')[0]
df = pd.read_excel(file,header=2)
df.head()
```

```python
df_list = []
for file in glob(r'C:\workspace\VScode\crawling\dain\주유소/지역*.xls'):
    gs = pd.read_excel(file, header=2)
    df_list.append(gs)
df = pd.concat(df_list)
df
```

```python
df = df[['상호', '주소', '휘발유','셀프여부','상표']]
df.head()
df.rename(columns={'휘발유':'가격', '셀프여부':'셀프'}, inplace=True)
df.head()
```

```python
df['구'] = df.주소.str.split().str[1]
df.head()
```

```python
len(df['구'].unique())

df.info()
```

```python
# 가격 '-'
df['가격'].unique()
```

```python
df = df[df.가격 != '-']
```

```python
df.가격 = df.가격.astype(int)
df.info()
```

```python
df.to_csv('../daou/서울_주유소(230418)', index=False)
```
