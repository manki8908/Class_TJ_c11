```python
import requests
from bs4 import BeautifulSoup
import re  # 정규표현식
import pandas as pd
headers = {'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/63.0.3239.132 Safari/537.36'}
base="https://www.genie.co.kr/chart/top200?ditc=D&ymd=20230417&hh=15&rtm=Y&pg="
pg = ['1','2','3','4']
```

```python
line = []

for i in range(len(pg)):

    url = base + pg[i]
    html = requests.get(url, headers = headers).text
    soup = BeautifulSoup(html, 'html.parser')

    bestList = soup.select('tbody > tr')

    for j in range(len(bestList)):

        try:
            rank = int(bestList[j].find('td', 'number').contents[0].strip())
        except:
            rank = -999

        try:
            subj = bestList[j].find('td', 'info').find('a', 'title').contents[-1].strip()
        except:
            subj = "---"
        
        try:
            arts = bestList[j].find('td', 'info').find('a', 'artist').contents[-1].strip()
        except:
            arts = "---"
        
        try:
            albm = bestList[j].find('td', 'info').find('a', 'albumtitle').contents[-1].strip()
        except:
            albm = '---'
        #print(albm)

        line.append([rank, subj, arts, albm])


df = pd.DataFrame(line, columns=['순위', '제목', '아티스트', '앨범'])

df


```
