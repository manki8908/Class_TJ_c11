```python
import requests
from bs4 import BeautifulSoup
import re  # 정규표현식
import pandas as pd
from tqdm.notebook import tqdm
headers = {'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/63.0.3239.132 Safari/537.36'}
base_url = "https://book.interpark.com"
url_main = base_url + "/display/collectlist.do?_method=bestsellerHourNew&bookblockname=b_gnb&booklinkname=%BA%A3%BD%BA%C6%AE%C1%B8&bid1=w_bgnb&bid2=LiveRanking&bid3=main&bid4=001"
```

```python
# .. main 링크 페이지
html_main = requests.get(url_main, headers = headers).text
soup_main = BeautifulSoup(html_main, 'html.parser')
soup_main
```

```python
line = []

bestList = soup_main.select('ol > li')

for j in tqdm(range(len(bestList))):
#for j in range(1):
    # rank
    rank_c = ''
    rank = bestList[j].select('.rankBtn_ctrl')
    for span in rank:
        rank_c += str(span['class'][1][-1])
    rank_c
    rank_n = int(rank_c)
    rank_n

    # name
    name = bestList[j].select_one('.itemName').text.strip()
    name

    # 가격
    price = bestList[j].find('em').text.strip()
    price_n = int(price.replace(',',""))
    price_n

    # .. 서브페이지 링크
    url_sub = base_url + bestList[j].select_one('div.coverImage label a')['href']

    html_sub = requests.get(url_sub, headers = headers).text
    soup_sub = BeautifulSoup(html_sub, 'html.parser')
    #soup_sub

    bestList_sub = soup_sub.select('ul.bInfo_txt > li')

    # 저, 역, 감수, 출판, 발행, 쪽수
    author, compan, transl, public, pagenb = '---', '---', '---', '---', '---'

    for i in range(len(bestList_sub)):

        #print(bestList_sub[i].text.split(':')[0], bestList_sub[i].text.split(':')[1])

        if (bestList_sub[i].text.split(':')[0].strip() == '저'):
            author = bestList_sub[i].text.split(':')[1]       
        if (bestList_sub[i].text.split(':')[0].strip() == '역'):
            transl = bestList_sub[i].text.split(':')[1]

        if (bestList_sub[i].text.split(':')[0].strip() == '발행'):
            public = bestList_sub[i].text.split(':')[1]

        if (bestList_sub[i].text.split(':')[0].strip() == '출판사'):
            compan = bestList_sub[i].text.split(':')[1]

        if (bestList_sub[i].text.split(':')[0].strip() == '쪽수'):
            pagenb = bestList_sub[i].text.split(':')[1]

        # 강사님 version
        #for li in bestList_sub:
        #    text = li.get_text().strip()
        #    if '발행 :' in text:
        #        date = text.split(':')[1].strip()
        #        date = re.sub('[^0-9]','', date)

    line.append([rank_n, name, price_n, author, transl, compan, public, pagenb])


df = pd.DataFrame(line, columns=['순위', '제목', '가격', '저자', '역자', '출판사', '발행일', '쪽수'])
df
```

```python
df.to_csv('../daou/인터파크서브out')
```
