```python
import numpy as np
import pandas as pd
import time
import re 
from tqdm.notebook import tqdm

import requests
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
```

```python
# 메인 홈페이지
url='https://search.danawa.com/dsearch.php?k1=%EB%AC%B4%EC%84%A0%EC%B2%AD%EC%86%8C%EA%B8%B0&module=goods&act=dispMain'
driver = webdriver.Chrome()
driver.get(url)
time.sleep(2)
```

```python
# page
p = [ str(i+1) for i in range(10) ]
line = []
```

```python
for j in range(len(p)):  # page loop
#for j in [0]:  # page loop

    # station 정보 추출
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    #print(soup)

    stn = soup.select('#productListArea div.main_prodlist > ul.product_list > li.prod_item > div.prod_main_info')

    #for i in [0]:
    for i in range(len(stn)):
        print("-------------", i)
        # 상품명
        name = stn[i].select_one('p.prod_name').text.strip()
        print(name)
        # 스펙
        addr = stn[i].select_one('div.spec_list').text.strip()
        spec = addr.replace(" ","").replace("\t","").replace("\n","")
        print(spec)
        # 가격
        price = stn[i].select_one('p.price_sect strong').text.strip()
        print(price)

        line.append([name, spec, price])

    time.sleep(2)
    
    if j <= 8:
        next_btn = next_p_btn1 = driver.find_element(by=By.XPATH, value=f'//*[@id="paginationArea"]/div/div/a[{j+2}]')
        next_btn.click()
```

```python
df = pd.DataFrame(line, columns=['품명', '스펙', '가격'])
df
```

```python
# 상품정보 태그에서 원하는 정보를 추출하는 함수
def get_prod_item(find_ele, num_page):

    base = "https://search.danawa.com/dsearch.php"
    find = f'?query={find_ele}&originalQuery={find_ele}&previousKeyword={find_ele}&volumeType=allvs&page={str(num_page)}&limit=40&sort=saveDESC&list=list&boost=true&addDelivery=N&mode=simple&recommendedSort=Y&defaultUICategoryCode=14237761&defaultPhysicsCategoryCode=1840%7C59152%7C59189%7C0&defaultVmTab=3014&defaultVaTab=968571&tab=goods'
    url= base + find
    driver = webdriver.Chrome()
    driver.get(url)
    time.sleep(2)

    line = []
    p = [ str(i+1) for i in range(num_page) ]

    for j in range(len(p)):  # page loop
    #for j in [0]:  # page loop

        # station 정보 추출
        soup = BeautifulSoup(driver.page_source, 'html.parser')
        #print(soup)

        stn = soup.select('#productListArea div.main_prodlist > ul.product_list > li.prod_item > div.prod_main_info')

        #for i in [0]:
        for i in range(len(stn)):
            print("-------------", i)
            # 상품명
            name = stn[i].select_one('p.prod_name').text.strip()
            print(name)
            # 스펙
            addr = stn[i].select_one('div.spec_list').text.strip()
            spec = addr.replace(" ","").replace("\t","").replace("\n","")
            print(spec)
            # 가격
            price = stn[i].select_one('p.price_sect strong').text.strip()
            print(price)

            line.append([name, spec, price])

        time.sleep(2)
    
        if j <= 8:
            next_btn = next_p_btn1 = driver.find_element(by=By.XPATH, value=f'//*[@id="paginationArea"]/div/div/a[{j+2}]')
            next_btn.click()

        df = pd.DataFrame(line, columns=['품명', '스펙', '가격'])
        
    return df
```
