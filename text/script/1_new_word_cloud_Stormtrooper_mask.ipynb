```python
# NLTK (natual Language Toolkit)
import nltk
#!pip install wordcloud
from wordcloud import WordCloud, STOPWORDS
from PIL import Image  # 그림 불러오는 패키지
import numpy as np
import matplotlib.pyplot as plt
#NLTK Stopwords
from nltk.corpus import stopwords
```

```python
# Alice.txt 저작권 주의
with open(r'C:\Users\tjoeun\Desktop\수업자료\수업_실습\DATA\text\A_new_hope.txt') as file:  
    text = file.read()

text = text.replace('HAN', "Han").replace("LUKES'S", "Luke")
```

```python
sw_wc = STOPWORDS
# said를 stopwords에 추가 -> 책에 said가 많이 나옴
sw_wc.add('int')
sw_wc.add('ext')
print(len(sw_wc))
```

```python
# 워드 크라우드 생성
wc = WordCloud(background_color='white',
               max_words=2000, stopwords=sw_wc)
wc = wc.generate(text)

# 단어 빈도수의 전체 비율 확인, 상위 10개까지
count = 0
for key, val in wc.words_.items():
    print(key, ':', val)
    count += 1
    if count > 10:        break
```

```python
#그림 위에 표시하기
mask = np.array(Image.open(r'C:\Users\tjoeun\Desktop\수업자료\수업_실습\DATA\text\Stormtrooper_mask.png'))
# seamap 에서 디폴트로 노란색, 보라색으로 설정되어 나와 다시 흑백처리
plt.imshow(mask, cmap=plt.cm.gray, interpolation='bilinear')  
plt.axis('off')
plt.show()
```

```python
wc = WordCloud(background_color='white', mask=mask,
               max_words=1000, stopwords=sw_wc).generate(text)
plt.figure(figsize=(10,12))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```
