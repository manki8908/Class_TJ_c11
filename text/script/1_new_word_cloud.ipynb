# 2. Alice in Wonderland

```python
# Alice.txt 저작권 주의
with open(r'C:\Users\tjoeun\Desktop\수업자료\수업_실습\DATA\text\Alice.txt') as file:  
    text = file.read()
```

```python
# said를 stopwords에 추가 -> 책에 said가 많이 나옴
sw_wc.add('said')
print(len(sw_wc))
```

```python
# 워드 크라우드 생성
wc = WordCloud(background_color='white',
               max_words=2000, stopwords=sw_wc)
wc = wc.generate(text)
text
```

```python
# 단어 빈도수의 전체 비율 확인, 상위 10개까지
count = 0
for key, val in wc.words_.items():
    print(key, ':', val)
    count += 1
    if count > 10:        break
```

```python
plt.figure(figsize=(10,6))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
#그림 위에 표시하기
mask = np.array(Image.open(r'C:\Users\tjoeun\Desktop\수업자료\수업_실습\DATA\text\Alice_mask.png'))
# seamap 에서 디폴트로 노란색, 보라색으로 설정되어 나와 다시 흑백처리
plt.imshow(mask, cmap=plt.cm.gray, interpolation='bilinear')  
plt.axis('off')
plt.show()
```

```python
wc = WordCloud(background_color='white', mask=mask,
               max_words=1000, stopwords=sw_wc).generate(text)
              
```

```python
plt.figure(figsize=(10,12))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

# 3. StarWars

```python
with open('A_new_hope.txt') as file:
    text = file.read()
text = text.replace('HAN', 'Han').replace("LUKE'S", 'Luke')
# HAN -> Han
# LUKE'S -> Luke
```

```python
# stop words에 int와 ext
sw = STOPWORDS
sw.add('int')
sw.add('ext')
len(sw)
```

```python
mask = np.array(Image.open('Stormtrooper_mask.png'))
plt.imshow(mask, cmap= plt.cm.gray, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
wc = WordCloud(mask=mask, max_words=1000, stopwords=sw, margin=10, random_state=2023).generate(text)
default_color = wc.to_array
```

```python
#HSL Hue, Satuation lightness
def grey_color(word, font_size, position, orientation, random_state, **kwargs):
    return f'hsl(0, 0%, {np.random.randint(60, 101)}%)'
```

```python
plt.figure(figsize=(12,12))
plt.imshow(wc.recolor(color_func=grey_color,random_state=2023), interpolation='bilinear')
plt.axis('off')
plt.show()
```
